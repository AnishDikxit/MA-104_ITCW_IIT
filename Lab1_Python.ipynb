## 🎲 The 10-Line Casino Challenge

**The Rules:**
1. The computer will pick a random number between 1 and 10.
2. You have exactly **3 attempts** to guess the number.
3. If you guess correctly, the program prints "You win!" and stops.
4. If you guess incorrectly, the program tells you if you are "Too high" or "Too low".
5. If you run out of attempts, the program prints "Game Over" and reveals the number.

*Hint: `input()` always grabs text (Strings). If you need an integer, you must wrap it in `int()` like this: `guess = int(input("Enter number: "))`*

```python
import random

# 1. The computer picks a random number between 1 and 10
secret_number = random.randint(1, 10)
print("Welcome to the Casino! I am thinking of a number between 1 and 10.")

for attempt in range(3):

    guess1=int((input("Guess the number :")))
    if(int(guess1)==secret_number):
        print("You WIN!")
        break
    elif guess1>secret_number:
        print("Too HIGH!")
    else :
        print("Too LOW!")

else:
    print(f"Game Over! The secret number was {secret_number}")
```
